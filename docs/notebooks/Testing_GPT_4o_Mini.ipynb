```python
from langchain_openai import ChatOpenAI
from google.colab import userdata
from rich import print

gpt4o_mini = ChatOpenAI(model = "gpt-4o-mini",
                        openai_api_key = userdata.get("OPENAI_API_KEY")
)
```

```python
## Mathematics
## Evaluates the model's capability to perform calculations and apply mathematical concepts to solve problems.

response = gpt4o_mini.invoke("""
If 20 shirts take 10 hours to dry. How much time will it take to dry 100 shirts?
""")


print(response.content)
```

```python
## Physical Reasoning
## Tests the model's ability to understand and predict outcomes in physical scenarios, often involving spatial relationships and object interactions.

response = gpt4o_mini.invoke("""Imagine you have a book, 9 eggs, a laptop, a bottle, and a nail.
How would you stack these objects in a stable manner?
Explain your reasoning for the order and placement of each item.""")
print(response.content)
```

```python
## Code Generation
## Assesses the model's ability to understand programming concepts and generate functional code based on given requirements.

response = gpt4o_mini.invoke("""
Create a snake game in Python in pygame
""")

print(response.content)
```

```python
## Theory of Mind
## Tests the model's capacity to understand and reason about the mental states, beliefs, and emotions of others.

response = gpt4o_mini.invoke("""John and Mary are at a party.
John sees Mary talking to another guest and feels jealous.
Why might John feel this way?""")

print(response.content)
```

```python
## Creative Writing
## Assesses the model's capacity to generate original, imaginative content across various literary forms and styles.

response = gpt4o_mini.invoke("""Write a short story of exactly 100 words that includes these elements:
a time traveler, a talking plant, and a moral dilemma. The story should end with an unexpected twist.
""")

print(response.content)
```
